```python
import random

def generate_numbers():
    numbers = []
    numbers.append(random.randint(1, 40))
    numbers.append(random.randint(2, 52))
    numbers.append(random.randint(12, 70))
    numbers.append(random.randint(24, 81))
    numbers.append(random.randint(36, 87))
    numbers.append(random.randint(60, 90))
    return numbers

def check_criteria(numbers):
    # Check odd numbers count
    odd_count = sum(1 for num in numbers if num % 2 != 0)
    if odd_count != 2 and odd_count != 4:
        return False

    # Check if second and third or third and fourth are in the same tens
    second_tens = numbers[1] // 10
    third_tens = numbers[2] // 10
    fourth_tens = numbers[3] // 10
    if (second_tens != third_tens) and (third_tens != fourth_tens):
        return False

    # Check differences between numbers
    if numbers[1] - numbers[0] <= 38:
        return False
    if numbers[2] - numbers[1] <= 39:
        return False
    if numbers[3] - numbers[2] <= 40:
        return False
    if numbers[4] - numbers[3] <= 38:
        return False
    if numbers[5] - numbers[4] <= 35:
        return False

    # Check if one number has the same ones digit
    ones_digit = [num % 10 for num in numbers]
    if ones_digit.count(ones_digit[0]) != 1:
        return False

    # Check if numbers are in ascending order
    if numbers != sorted(numbers):
        return False

    # Check total range
    total = sum(numbers)
    if total <= 170 or total >= 370:
        return False

    return True

# Generate valid rows of numbers
valid_rows = []
while len(valid_rows) < 40:
    numbers = generate_numbers()
    if check_criteria(numbers):
        valid_rows.append(numbers)

# Save the results in a text file
with open('goldlotux_results.txt', 'w') as file:
    for row in valid_rows:
        file.write(' '.join(f'{num:02d}' for num in row))
        file.write('\n')

print("Results saved to 'results.txt'.")
```
